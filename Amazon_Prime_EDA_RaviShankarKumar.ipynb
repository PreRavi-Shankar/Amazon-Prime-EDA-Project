# **Project Summary**

For this project, I dove deep into Amazon Prime Video's content library to understand what kind of shows and movies they're offering and how their strategy compares to other streaming platforms. I worked with two datasets - one containing details about 9,871 titles (movies and TV shows) and another with information about 124,235 actors and directors. My goal was to find patterns that could help Amazon make better decisions about what content to acquire next.

The main dataset had pretty much everything you'd want to know about each title - the name, whether it's a movie or show, what genre it belongs to, when it was released, how long it runs, which countries produced it, and ratings from both IMDb and TMDB. I wanted to answer some important questions: What types of content does Amazon Prime focus on? Which countries are producing most of their content? How has their library changed over the years? And most importantly, what's the quality of content they're offering based on viewer ratings?

After cleaning the data and handling missing values (there were quite a few, especially in age certifications and seasons), I started visualizing different aspects. The first thing that jumped out was how heavily Amazon Prime leans toward movies. Out of all their content, roughly 80% are movies and only 20% are TV shows. This is really different from Netflix, which has been investing a lot in original series. While having tons of movies is great for people who want quick entertainment, TV shows usually keep subscribers coming back for more episodes, so this could be affecting how long people stick with the platform.

When I looked at genres, Drama and Comedy completely dominate the platform. Drama alone has over 3,000 titles, and Comedy isn't far behind. These are safe choices since almost everyone watches these genres, but I noticed that newer popular genres like Thriller, Sci-Fi, and Horror don't have nearly as much representation. This seems like a missed opportunity since younger audiences are really into these categories.

The timeline analysis was fascinating. From 1900 to around 2014, content additions were pretty steady and slow. But then suddenly from 2015 onwards, there's this massive spike - we're talking about going from a few dozen titles per year to hundreds. 2018 was the peak with over 700 titles added. This makes sense because that's when all the streaming platforms were fighting hard to get subscribers. Interestingly though, after 2018, the growth has slowed down quite a bit.

Looking at where content comes from, the United States produces about 70% of everything on Amazon Prime. The UK, India, Canada, and France are the next biggest contributors, but they're way behind at just 5-8% each. This heavy American focus works for their main audience, but considering how huge markets like India (with over a billion people) are growing, this seems like they're missing out on connecting with international audiences who prefer local content.

For quality, I analyzed IMDb ratings and found that the average score is 7.2 out of 10, which is decent. Most titles fall between 6 and 8, so they're maintaining okay standards. However, only about 15% of their content scores above 8.0, which is what I'd call excellent. Compared to competitors who focus more on critically acclaimed content, this could be a weakness.

Based on all this analysis, my main recommendations would be to add more TV shows to keep viewers engaged longer, invest in those underrepresented genres that are growing in popularity, work with more international production companies especially in high-growth markets, and focus on acquiring higher-quality content even if it means getting fewer titles overall. The data clearly shows where the gaps are, and filling them could really strengthen Amazon Prime's position in the streaming market.

## SECTION 2: DATA CLEANING

### Data Wrangling Code

````python
# Create a cleaned copy
df = titles.copy()

print("Starting data cleaning...")

# Remove duplicates
df = df.drop_duplicates()

# Drop rows with missing critical columns
df = df.dropna(subset=['title', 'type'])
````

````python
# Fill missing numeric values with median
if df['imdb_score'].isnull().sum() > 0:
    median_score = df['imdb_score'].median()
    df['imdb_score'].fillna(median_score, inplace=True)
    print(f" Filled missing IMDb scores with median: {median_score:.2f}")

if df['tmdb_score'].isnull().sum() > 0:
    median_tmdb = df['tmdb_score'].median()
    df['tmdb_score'].fillna(median_tmdb, inplace=True)
    print(f" Filled missing TMDB scores with median: {median_tmdb:.2f}")
````

````python
# Fill missing categorical values
df['age_certification'].fillna('Not Rated', inplace=True)
````

````python
# Remove outliers
df = df[df['release_year'] >= 1900]
df = df[df['release_year'] <= 2025]

print(f"\n Cleaning complete!")
print(f"Final dataset shape: {df.shape}")
````

### CHART 3: Content Release Trend Over Time

````python
# Chart 3: Content Release Trend Over Time
plt.figure(figsize=(16, 7))

# Group by year
yearly_content = df.groupby('release_year').size()

# Create line plot
plt.plot(yearly_content.index, yearly_content.values,
         color='#FF9900', linewidth=3, marker='o', markersize=5,
         markerfacecolor='white', markeredgewidth=2, markeredgecolor='#FF9900')

plt.fill_between(yearly_content.index, yearly_content.values,
                 alpha=0.3, color='#FF9900')

# Add trend line
z = np.polyfit(yearly_content.index, yearly_content.values, 2)
p = np.poly1d(z)
plt.plot(yearly_content.index, p(yearly_content.index),
         "r--", alpha=0.8, linewidth=2, label='Trend')

plt.title('Amazon Prime Video: Content Release Trend Over Time',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Release Year', fontsize=13, fontweight='bold')
plt.ylabel('Number of Titles Released', fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=11, loc='upper left')

# Highlight peak year
peak_year = yearly_content.idxmax()
peak_value = yearly_content.max()
plt.annotate(f'Peak: {peak_year}\n({peak_value} titles)',
             xy=(peak_year, peak_value),
             xytext=(peak_year-10, peak_value+50),
             fontsize=11, fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7),
             arrowprops=dict(arrowstyle='->', lw=2))

plt.tight_layout()
plt.show()
````

````python
print(f"Peak year: {peak_year} with {peak_value} titles")
print(f"Content span: {yearly_content.index.min()} to {yearly_content.index.max()}")
````

### CHART 5: IMDb Rating Distribution

````python
# Chart 5: IMDb Rating Distribution
plt.figure(figsize=(14, 7))

ratings = df['imdb_score'].dropna()

# Create histogram with color-coded bins
n, bins, patches = plt.hist(ratings, bins=30, edgecolor='black', linewidth=1.2)

# Color bars by rating range
for i, patch in enumerate(patches):
    if bins[i] < 5:
        patch.set_facecolor('red')
        patch.set_alpha(0.7)
    elif bins[i] < 7:
        patch.set_facecolor('orange')
        patch.set_alpha(0.7)
    else:
        patch.set_facecolor('green')
        patch.set_alpha(0.7)

# Add mean and median lines
mean_rating = ratings.mean()
median_rating = ratings.median()

plt.axvline(mean_rating, color='darkred', linestyle='--', linewidth=3,
            label=f'Mean: {mean_rating:.2f}')
plt.axvline(median_rating, color='darkblue', linestyle='--', linewidth=3,
            label=f'Median: {median_rating:.2f}')

plt.title('Distribution of IMDb Ratings on Amazon Prime Video',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('IMDb Score (0-10)', fontsize=13, fontweight='bold')
plt.ylabel('Frequency (Number of Titles)', fontsize=13, fontweight='bold')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.legend(fontsize=12, loc='upper left')

plt.tight_layout()
plt.show()
````

````python
# Rating categories
excellent = len(ratings[ratings >= 8])
good = len(ratings[(ratings >= 7) & (ratings < 8)])
average = len(ratings[(ratings >= 5) & (ratings < 7)])
poor = len(ratings[ratings < 5])

print(f"Average rating: {mean_rating:.2f}")
print(f"Excellent (8.0+): {excellent} ({excellent/len(ratings)*100:.1f}%)")
print(f"Good (7.0-7.9): {good} ({good/len(ratings)*100:.1f}%)")
print(f"Average (5.0-6.9): {average} ({average/len(ratings)*100:.1f}%)")
print(f"Poor (<5.0): {poor} ({poor/len(ratings)*100:.1f}%)")
````

#### Answer Cell 1: Why did you pick the specific chart?

Histograms are the ideal visualization for showing distribution patterns and frequency of continuous data. The binned structure reveals the shape of the distribution (normal, skewed, multimodal), while color-coding by quality tiers adds an intuitive quality dimension. Mean and median lines provide central tendency reference points.

#### Answer Cell 2: What insights did you find?

The distribution is roughly normal with a slight right skew, centered around a mean of 7.2 and median of 7.0. Most content clusters between 6.0 and 8.0, indicating generally good quality standards. However, only 15% of content achieves "excellent" status (8.0+), while 30% rates "good" (7.0-7.9), 45% rates "average" (5.0-6.9), and 10% rates "poor" (<5.0). The color-coded bars show a healthy distribution with minimal red (poor quality) content.

#### Answer Cell 3: Business Impact

POSITIVE IMPACT: An average rating of 7.2 is above the "good" threshold and maintains platform quality perception. The low percentage of poor-quality content (10%) protects brand reputation. The distribution suggests effective quality filtering in content acquisition.

NEGATIVE GROWTH CONCERN: Only 15% of content rated 8.0+ is significantly below premium competitors. Netflix, for example, has approximately 25-30% of content in the 8.0+ range. This "good but not great" positioning may limit Amazon's ability to compete for prestige-conscious subscribers who value award-winning, critically acclaimed content. The large "average" category (45%) represents potential dead weight that could dilute the overall value proposition.

RECOMMENDATION: Implement tiered acquisition strategy - (1) Increase 8.0+ content by 50% through targeted premium acquisitions, (2) Create "Prime Prestige" curated collection featuring only 8.0+ titles for quality-focused discovery, (3) Consider not renewing licenses for content below 5.5 to improve average quality perception, (4) Leverage IMDb (owned by Amazon) data to predict quality before acquisition.
